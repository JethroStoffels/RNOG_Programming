```python
import numpy as np
from NuRadioReco.utilities import units
print("Old interpolation freqs")
L=np.concatenate((np.arange(10 * units.MHz,150 * units.MHz,20 * units.MHz),np.arange(150 * units.MHz,1200 * units.MHz,200 * units.MHz)))
for f in L:
    print(np.round(f/units.MHz))
print(10*"=")
print("New interpolation freqs")
L=np.arange(50 * units.MHz,150 * units.MHz,10 * units.MHz)
for f in L:
    print(np.round(f/units.MHz))
```

```python
StNr,Date,ChNrs=23,datetime(2022, 8, 15),[13,16,19]
#jsons=["RNO_season_2022.json","RNO_season_2022+30.json","RNO_season_2022-phi+30.json","RNO_season_2022-phi-30.json"]
jsons=["RNO_season_2022-phi-30.json"]
```

```python
for detpath in [os.path.dirname(detector.__file__) + "/RNO_G/" + f for f in jsons]:
    GNDetector = detector.Detector(json_filename = detpath)#,antenna_by_depth=False)
    GNDetector.update(Date)
    GNStation=station.Station(StNr)
    print(detpath)
    antenna_pattern_provider=NuRadioReco.detector.antennapattern.AntennaPatternProvider()
    for ChNr in ChNrs:
        GNChannel=NuRadioReco.framework.channel.Channel(ChNr)
        GNChannel.set_trace(trace=np.zeros(2048), sampling_rate=GNDetector.get_sampling_frequency(StNr, ChNr))
        GNStation.add_channel(GNChannel) 
        del GNChannel
    for channel in GNStation.iter_channels():
        print(channel.get_id())
        freqs = channel.get_frequencies()
        antenna_pattern = antenna_pattern_provider.load_antenna_pattern(GNDetector.get_antenna_model(GNStation.get_id(), channel.get_id()))
        antenna_orientation = GNDetector.get_antenna_orientation(GNStation.get_id(), channel.get_id())
        print("[ori_theta, ori_phi, rot_theta,rot_phi]=",antenna_orientation)
        antenna_response = antenna_pattern.get_antenna_response_vectorized(freqs, 20, 90, *antenna_orientation)
        plt.figure()
        plt.title("St" +str(GNStation.get_id())+ "Ch" + str(channel.get_id()) + " for " + detpath.split("/")[-1])
        plt.plot(freqs,np.abs(antenna_response['theta']),label='theta')
        plt.plot(freqs,np.abs(antenna_response['phi']),label='phi')
        plt.legend()
        plt.savefig("Figures/AEVels/"+ "St" +str(GNStation.get_id())+ "Ch" + str(channel.get_id()) +"-"+ detpath.split("/")[-1][:-5])
        plt.show()
```
