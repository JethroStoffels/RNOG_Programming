# Time computations:

For the computation of the LST, the position of the detector is required. For Station 22 it is the following: <br>
Station 22: Longitude, Latitude  = -38.459936 , 72.598265 <br>
This can be approximated to remain constant between stations as they are only located 1.25 km away from eachother

```python
T=datetime.utcfromtimestamp(TriggerTimes[EvIdx])
print("Date/time: ",T.strftime('%Y-%m-%d %H:%M:%S'))
print("T.hour: ",T.hour)
print("T.minute: ",T.minute)
print("T.second: ",T.second)
print("Hour in decimals (manual): ",T.hour+T.minute/60 + T.second/3600)
print("Hour in decimals (automated): ",UTC(TriggerTimes,EvIdx))
print("LST time in decimal hours: ",LST(TriggerTimes,EvIdx))
```

# Transit curve plot:

In order to construct a transit curve, the $V_{RMS}$ value of each timetrace under study is calculated together with its time of occurrence in Local Sidereal Time. These $V_{RMS}$ values are then binned according to their LST time and the average $V_{RMS}$ value per bin is calculated.

```python
RunList=np.concatenate((np.arange(8,23),np.arange(91,111)))
for BadRun in [16,17,97,98]:
    RunList=np.delete(RunList,np.where(RunList==BadRun)[0][0])
```

```python
EventTime, EventRMS, MidBins, GroupedVRMS, FilteredRuns,FilteredEvNrs=TransitCurve(22,13,RunList,NBins=4*24,ZeroAvg=False,TimeFormat="LST",Triggers=(5,5,5,5),StdCut=(-1,-1), FFTFilter=False,Lowpass=False,HardCut=-1,EventSkip=25,Plot=True)
```
