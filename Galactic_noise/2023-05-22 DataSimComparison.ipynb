```python
RelErrorData=[[10, 9.913043478260875],[22.093386751530797, 8.43478260869566],[45.26241319388085, 10.260869565217394],[85.98569967363346, 5.478260869565222],[151.47043367743976, 5.65217391304348],[404.1637673000108, 6.0000000000000036],[812.5237525193901, 4],[1378.301606322275, 2.956521739130441],[1431.3231808832006, 5.391304347826093],[2338.042810564148, 10.086956521739133]]
RelError=[RelErrorData[i][1] for i in range(len(RelErrorData))]
RelErrorFreq=[RelErrorData[i][0] for i in range(len(RelErrorData))]
print(RelError)
print(RelErrorFreq)

import scipy.interpolate
import numpy as np
TNoise,Noise_max_freq,Noise_min_freq=300,1100 ,10
interpolation_frequencies=np.arange(Noise_min_freq, Noise_max_freq,100)

RelError_interpolator = scipy.interpolate.interp1d(RelErrorFreq, RelError,kind="linear")
RelError_interpolated =RelError_interpolator(interpolation_frequencies)

import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.plot(RelErrorFreq,RelError,label="Relative error")
plt.plot(interpolation_frequencies,RelError_interpolated,'o',label="100MHz spacing interpolation")
plt.xscale('log')
plt.ylabel('Relative error %')
plt.xlabel("Frequency (MHz)")
plt.legend()
plt.show()
```
