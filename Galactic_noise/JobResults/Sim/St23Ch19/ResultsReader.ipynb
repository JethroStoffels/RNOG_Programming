```python
def CombineResults(ClusterName1,ClusterName2):
    import numpy as np
    import os
    SimEventRMSTot=np.array([])
    SimEventTimeTot=np.array([])
    #for i in np.delete(range(242),[102]):
    SimEventRMSTot=np.append(SimEventRMSTot,np.load("HCSimEventRMS_" + ClusterName1 + "-" + "Tot" + ".npy",allow_pickle=True))
    SimEventTimeTot=np.append(SimEventTimeTot,np.load("HCSimEventTime_" + ClusterName1 + "-" + "Tot" + ".npy",allow_pickle=True))

    SimEventRMSTot=np.append(SimEventRMSTot,np.load("SSimEventRMS_" + ClusterName2 + "-" + "Tot" + ".npy",allow_pickle=True))
    SimEventTimeTot=np.append(SimEventTimeTot,np.load("SSimEventTime_" + ClusterName2 + "-" + "Tot" + ".npy",allow_pickle=True))
    
    np.save('CEventRMS_' + "Tot", SimEventRMSTot)
    np.save('CEventTime_' +  "Tot", SimEventTimeTot)
    return SimEventRMSTot, SimEventTimeTot
```
